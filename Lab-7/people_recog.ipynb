```python
import pandas as pd
import cv2

import os
```

```python
bboxes = pd.read_csv("./PRW-v16.04.20/query_info.txt", sep=' ', names=['id', 'x', 'y', 'w', 'h', 'image'])
bboxes.sort_values(by='image')
```

```python
import shutil

images_path = './PRW-v16.04.20/frames/'

for img_name in os.listdir(images_path):
    img_path = os.path.join(images_path, img_name)
    img = cv2.imread(img_path)
    w = img.shape[0]
    h = img.shape[1]

    img_name_clean = img_name.split('.')[0]
    boxes = bboxes[bboxes['image'] == img_name_clean]
    if boxes.empty:
        path = "./PRW-v16.04.20 rebuilded/empty"
    else:
        path = "./PRW-v16.04.20 rebuilded/people"
        
    with open(os.path.join(path, f"labels/{img_name_clean}.txt"), "w") as file:
        for box in boxes.values:
            # 0 - если есть человек, пустой файл - нет человека
            # print(box)
            file.write(f"0 {float(box[1]) / w} {float(box[2]) / h} {float(box[3]) / w} {float(box[4]) / h}\n")

    shutil.move(img_path, os.path.join(path, f"images/{img_name}"))
```

```python
train_part = 0.66
train_val_part = 0.85

path_empty = "./PRW-v16.04.20 rebuilded/empty/images/"
path_people = "./PRW-v16.04.20 rebuilded/people/images/"

n_files_empty = len(os.listdir(path_empty))
n_files_people =  len(os.listdir(path_people))

print("Всего изображений пустых: ", n_files_empty)
print("Всего изображений с людьми: ", n_files_people)
```

```python
train_part_n = int(n_files_empty*train_part)
train_val_part_n = int(n_files_empty*train_val_part)

path_new = "./PRW-v16.04.20 rebuilded/complete/"
for img_name in os.listdir(path_empty):
    shutil.copy(path_empty, os.path.join(path_new, f"images/{img_name}"))
    shutil.copy(path_empty, os.path.join(path_new, f"labels/{img_name}"))
    shutil.copy(path_empty, os.path.join(path_new, f"images/{img_name}"))
    shutil.copy(path_empty, os.path.join(path_new, f"labels/{img_name}"))
```
